```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 🔹 Load Qwen model (lightweight + good chat)
model_name = "Qwen/Qwen2.5-0.5B-Instruct"

print("Loading model... Please wait ⏳")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32,   # CPU friendly
    device_map="auto"
)

# 🔹 Store conversation history
chat_history = []

print("Chatbot: Hello! I am your AI assistant. Type 'exit' to end the chat.")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! 👋")
        break

    # 🔹 Add user input to history
    chat_history.append(f"User: {user_input}")

    # 🔹 Keep last 5 messages (memory control)
    history = "\n".join(chat_history[-5:])

    # 🔹 Strong chat prompt
    prompt = f"""
You are a helpful, intelligent AI assistant.
Give clear, accurate, and non-repetitive answers.
Keep responses short (2-3 sentences).

Conversation:
{history}

Assistant:
"""

    # 🔹 Tokenize
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # 🔹 Generate response
    outputs = model.generate(
        **inputs,
        max_new_tokens=120,
        do_sample=True,          # enable natural responses
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.2   # reduce repetition
    )

    # 🔹 Decode response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # 🔹 Extract only assistant answer
    response = response.split("Assistant:")[-1].strip()

    print("Chatbot:", response)

    # 🔹 Save response to history
    chat_history.append(f"Assistant: {response}")
```
